```python
import time
import random

#Вводим массив....для хранения значений затраченных количеств попыток на угадывания числа, ед. измерения = 1 игра
array_trues = []
#.......массив для хранения значений затраченного времени на 1 игру
array_times = []
#Сюда запишем среднее значение среди всего массива
average_value = None


#.....Функция-декоратор, используем для цикличного запуска функции-игры, и записи полученных результатов в массив
def calc_average(function):
    global array_trues
    global array_times
    global average_value
    count_average = 0      #Переменная счетчик для функции-декоратора, выборка 1000 игр

    #.....Запускаем цикл, и проигрываем игру-функцию 1000 раз.
    while True:
        if count_average < 1000:
            function()
            count_average += 1
        else:
            break

    #....Показываем расчет по "среднемуо количеству попыток угадать" и "зтраченному времени на угадывание", ед. измерения 1 игра
    average_value = round(sum(array_trues) / 1000)
    print(f'Среднее количество попыток угадать: {average_value}')
    print('Выборка....')
    time.sleep(2)
    print('Среди 1000 игр\n')
    time.sleep(2)
    print(f'Компьютер потратил в среднем на 1-ну игру: {(sum(array_times) / 10):0.8f} сек.')

#.....Функция-игра, для генерации числа от 1 до 100, и алгоритма угадывания для компьютера.
def randomizer():
    #  - используем для генерации числа
    random_number = random.randint(1, 100)
    #  - для сравнения с random_number....Будем использовать чтобы возвращать True для прерывания цикла.
    shadow_number = None   
    #  -  Счетчик для подсчета попыток, в конце игры перемещается в массив для подсчета среднего значения
    count_trues = 0                      

    #...Засекаем время, и начинаме цикл.
    start_time = time.perf_counter()
    while random_number != shadow_number:
        #Первая отправная точка...Для дальнейших ответвлений при random_number != 50
        shadow_number = 100 * 0.5
        count_trues += 1

        if random_number > 100 * 0.5:
            #....Для дальнейших ответлений при random_number != 75
            shadow_number = 100 * 0.75
            count_trues += 1

            if 100 * 0.5 < random_number < 100 * 0.75:
                for i in range(51, 75):
                    if i != random_number:
                        count_trues += 1
                    else:
                        random_number = i
                        shadow_number = i
                        count_trues += 1
                        break

            elif 100 * 0.75 < random_number <= 100:
                for i in range(76, 101):
                    if i != random_number:
                        count_trues += 1
                    else:
                        random_number = i
                        shadow_number = i
                        count_trues += 1
                        break

        if random_number < 100 * 0.5:
            #....Для дальнейших ответлений при random_number != 25
            shadow_number = 100 * 0.25
            count_trues += 1

            if 1 <= random_number < 100 * 0.25:
                for i in range(1, 25):
                    if i != random_number:
                        count_trues += 1
                    else:
                        random_number = i
                        shadow_number = i
                        count_trues += 1
                        break

            elif 100 * 0.25 < random_number < 100 * 0.5:
                for i in range(26, 50):
                    if i != random_number:
                        count_trues += 1
                    else:
                        random_number = i
                        shadow_number = i
                        count_trues += 1
                        break
    #...Заначиваем игру, переносим полученное количество попыток и время на игру в массивы
    end_time = time.perf_counter() - start_time
    array_trues.append(count_trues)
    array_times.append(end_time)

#Наслаждаемся нубокодом
calc_average(randomizer)
```
